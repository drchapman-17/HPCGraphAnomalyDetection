```python
import pandas as pd
from tqdm import tqdm
import numpy as np

df_list = []
for i in tqdm(range(49)):
    df_list.append(pd.read_parquet(f'rack_{i}.parquet').rename(columns={'values':f'values_{i}'}))

dfone = df_list[0]
for df in tqdm(df_list[1:]):
    dfone = pd.merge(dfone,df,how='outer',on='timestamp')
values_columns = [f'values_{i}' for i in range(49)]
dfone['values']= dfone[values_columns].agg(np.concatenate,axis=1)
dfone.drop(values_columns,axis=1)
```
